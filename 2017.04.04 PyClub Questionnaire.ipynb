# Multiples of 3 and 5

```python
total = 0
for i in range(3,10):
    if(i%3 ==0) or (i%5==0):
        total += i
print total
```

# PyClub Questionnaire

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
```

```python
df1 = pd.read_csv("C:/Users/hela.zouaghi/Downloads/PyClubData.csv")
df = df1.replace(np.nan, 'Unanswered', regex=True)
```

```python
df[[1]]
```

```python
a = df.groupby("What is your python level?")["What is your python level?"].count()
```

```python
sizes = a.values
```

```python
labels = a.index
colors = ['hotpink', 'lavender','turquoise','burlywood']
```

```python
plt.pie(sizes, labels=labels, colors=colors, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
cols = df.columns
for column_name in cols:
    a=df.groupby(column_name)[column_name].count()
    print column_name
    sizes = a.values
    labels = a.index
    plt.pie(sizes, labels=labels, colors=colors, autopct = '%1.1f%%')
    plt.axis('equal')
    plt.show()
```
